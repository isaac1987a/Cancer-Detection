```python
import keras
```

```python
# Developed with help from https://keras.io/api/models/model/
inputs = keras.Input(shape=(32,32,3))
def activationBlock(input, filters):
    filters = 32
    kernel_size=(3,3)
    conv1 = keras.layers.Conv2D(
        filters,
        kernel_size,
        strides=(1, 1),
        padding="valid",
        data_format=None,
        dilation_rate=(1, 1),
        groups=1,
        activation="sigmoid",
        use_bias=True,
        kernel_initializer="glorot_uniform",
        bias_initializer="zeros",
        kernel_regularizer=None,
        bias_regularizer=None,
        activity_regularizer=None,
        kernel_constraint=None,
        bias_constraint=None,
    ) (inputs)
    conv2 = conv1 = keras.layers.Conv2D(
        filters,
        kernel_size,
        strides=(1, 1),
        padding="valid",
        data_format=None,
        dilation_rate=(1, 1),
        groups=1,
        activation="sigmoid",
        use_bias=True,
        kernel_initializer="glorot_uniform",
        bias_initializer="zeros",
        kernel_regularizer=None,
        bias_regularizer=None,
        activity_regularizer=None,
        kernel_constraint=None,
        bias_constraint=None,
    ) (conv1)
    pool = keras.layers.MaxPooling2D(
        pool_size=(2, 2), strides=None, padding="valid", data_format=None, name=None
    )(conv2)
    dropout = keras.layers.Dropout(0.25)(conv2)
    return dropout
    
activation1 = activationBlock(inputs, 32)
activation2 = activationBlock(activation1, 32)
activation3 = activationBlock(activation2, 32)
activation4 = activationBlock(activation3, 32)
activation5 = activationBlock(activation4, 32)

dense1 = keras.layers.Dense(1024, activation='sigmoid') (activation5)
drop1 = keras.layers.Dropout(0.25)(dense1)
dense2 = keras.layers.Dense(1024, activation='sigmoid') (drop1)
drop2 = keras.layers.Dropout(0.25)(dense2)
dense3 = keras.layers.Dense(256, activation='sigmoid') (drop2)
drop3 = keras.layers.Dropout(0.25)(dense3)


outputs = keras.layers.Dense(10, activation="tanh")(drop3)

full_model = keras.Model(inputs, outputs)

# set up optomizer
optomizer = keras.optimizers.Adam(learning_rate=0.001)
keras.models.model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
```
